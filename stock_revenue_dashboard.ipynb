# Plot Tesla Stock Graph

```python
def make_graph(stock_data, revenue_data, stock):
    fig = make_subplots(rows=2, cols=1, shared_xaxes=True,
                        subplot_titles=(f"{stock} Stock Price", f"{stock} Revenue"),
                        vertical_spacing=0.3)

    stock_data_specific = stock_data[stock_data.Date <= '2021-06-14']
    revenue_data_specific = revenue_data[revenue_data.Date <= '2021-06-14']

    fig.add_trace(
        go.Scatter(x=stock_data_specific.Date, y=stock_data_specific.Close.astype("float"), name="Stock Price"),
        row=1, col=1
    )

    fig.add_trace(
        go.Scatter(x=revenue_data_specific.Date, y=revenue_data_specific.Revenue.astype("float"), name="Revenue"),
        row=2, col=1
    )

    fig.update_layout(showlegend=False, height=600, title=f"{stock} Stock Price and Revenue")
    fig.show()

# Plot Tesla data
make_graph(tesla_data, tesla_revenue, "Tesla")
```
